<h1> for timestamps </h1>
<li> probably should add a short format to UTC to get date, date time to seconds, data time to microseconds

```python
def Z(epoch_seconds=None):
  if epoch_seconds:
    iso = datetime.datetime.fromtimestamp(epoch_seconds).isoformat()
  else:
    iso = datetime.datetime.utcnow().isoformat()
#    iso =datetime.datetime.utcnow().strftime('%Y-%m-%dT%H:%M:%S')
  return iso + 'Z'
```
